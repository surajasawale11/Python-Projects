```python
from tkinter import *

def reset():
    entry_vadapav.delete(0,END)
    entry_tea.delete(0,END)
    entry_samosa.delete(0,END)
    entry_misal_pav.delete(0,END)
    entry_vada_usal.delete(0,END)
    entry_idli_sambhar.delete(0,END)
    entry_masala_dosa.delete(0,END)
    entry_bisleri.delete(0,END)
    entry_pav.delete(0,END)
    textarea.delete(1.0,END)
    
def total():
    try:a1=int(vadapav.get())
    except:a1=0
    try:a2=int(tea.get())
    except:a2=0
    try:a3=int(samosa.get())
    except:a3=0
    try:a4=int(misal_pav.get())
    except:a4=0
    try:a5=int(vada_usal.get())
    except:a5=0
    try:a6=int(idli_sambhar.get())
    except:a6=0
    try:a7=int(masala_dosa.get())
    except:a7=0
    try:a8=int(bisleri.get())
    except:a8=0
    try:a9=int(pav.get())
    except:a9=0
    c1=a1*15
    c2=a2*10
    c3=a3*15
    c4=a4*60
    c5=a5*60
    c6=a6*60
    c7=a7*60
    c8=a8*20
    c9=a9*10
    textarea.delete(1.0,END)
    textarea.insert(END,chars='-------------------------------------------------------\n')
    textarea.insert(END,chars="\n******************* WELCOME CUSTOMER ******************\n\n")
    textarea.insert(END,chars='-------------------------------------------------------\n')
    textarea.insert(END,chars="\tProducts\t\t Quantity\t\tPrice\n")
    if entry_vadapav.get()!="":
        textarea.insert(END,chars=f'\n\tVada Pav\t\t\t{entry_vadapav.get()}\t{c1} Rs\n')
    if entry_tea.get()!="":
        textarea.insert(END,chars=f'\n\tTea\t\t\t{entry_tea.get()}\t{c2} Rs\n')
    if entry_samosa.get()!="":
        textarea.insert(END,chars=f'\n\tSamosa\t\t\t{entry_samosa.get()}\t{c3} Rs\n')
    if entry_misal_pav.get()!="":
        textarea.insert(END,chars=f'\n\tMisal Pav\t\t\t{entry_misal_pav.get()}\t{c4} Rs\n')
    if entry_vada_usal.get()!="":
        textarea.insert(END,chars=f'\n\tVada Usal\t\t\t{entry_vada_usal.get()}\t{c5} Rs\n')
    if entry_idli_sambhar.get()!="":
        textarea.insert(END,chars=f'\n\tIdli Sambhar\t\t\t{entry_idli_sambhar.get()}\t{c6} Rs\n')
    if entry_masala_dosa.get()!="":
        textarea.insert(END,chars=f'\n\tMasala Dosa\t\t\t{entry_masala_dosa.get()}\t{c7} Rs\n')
    if entry_bisleri.get()!="":
        textarea.insert(END,chars=f'\n\tBisleri\t\t\t{entry_bisleri.get()}\t{c8} Rs\n')
    if entry_pav.get()!="":
        textarea.insert(END,chars=f'\n\tPav\t\t\t{entry_pav.get()}\t{c9} Rs\n')
    textarea.insert(END,chars='-------------------------------------------------------\n')
    totalbill=c1+c2+c3+c4+c5+c6+c7+c8+c9
    textarea.insert(END,chars=f'\n\tTotal Bill \t\t\t\t{totalbill} Rs\n')
    textarea.insert(END,chars='-------------------------------------------------------\n')

root=Tk()
root.geometry('1400x700')
root.title('Bill Management')
# root.resizable(False,False)

Label(text='Bill Management',bg='antiquewhite',fg='black',font=('calibari',38),width=300,height=2).pack()

f=Frame(root,bg='Gold',highlightbackground='black',highlightthickness=1,width=410,height=550)
f.place(x=10,y=125)

Label(f,text='Menu',font=('Gabriola',40,'bold'),fg='black',bg='Gold').place(x=10,y=0)
Label(f,font=('Lucida Calligraphy',15,'bold'),text='Vada Pav................Rs.15',fg='black',bg='Gold').place(x=10,y=80)
Label(f,font=('Lucida Calligraphy',15,'bold'),text='Tea.....................Rs.10',fg='black',bg='Gold').place(x=10,y=115)
Label(f,font=('Lucida Calligraphy',15,'bold'),text='Samosa..................Rs.15',fg='black',bg='Gold').place(x=10,y=150)
Label(f,font=('Lucida Calligraphy',15,'bold'),text='Misal Pav...............Rs.60',fg='black',bg='Gold').place(x=10,y=185)
Label(f,font=('Lucida Calligraphy',15,'bold'),text='Vada Usal...............Rs.60',fg='black',bg='Gold').place(x=10,y=220)
Label(f,font=('Lucida Calligraphy',15,'bold'),text='Idli Sambhar............Rs.60',fg='black',bg='Gold').place(x=10,y=255)
Label(f,font=('Lucida Calligraphy',15,'bold'),text='Masala Dosa.............Rs.60',fg='black',bg='Gold').place(x=10,y=295)
Label(f,font=('Lucida Calligraphy',15,'bold'),text='Bisleri.................Rs.20',fg='black',bg='Gold').place(x=10,y=330)
Label(f,font=('Lucida Calligraphy',15,'bold'),text='Pav.....................Rs.20',fg='black',bg='Gold').place(x=10,y=365)

f1=Frame(root,bd=5,height=550,width=600,relief=RAISED,bg='Chocolate')
f1.place(x=425,y=125)

vadapav=StringVar()
tea=StringVar()
samosa=StringVar()
misal_pav=StringVar()
vada_usal=StringVar()
idli_sambhar=StringVar()
masala_dosa=StringVar()
bisleri=StringVar()
pav=StringVar()

lbl_vadapav=Label(f1,font=('aria',20,'bold'),text='Vada Pav',width=20,fg='black',bg='Chocolate')
lbl_vadapav.grid(row=1,column=0)
lbl_tea=Label(f1,font=('aria',20,'bold'),text='Tea',width=20,fg='black',bg='Chocolate')
lbl_tea.grid(row=2,column=0)
lbl_samosa=Label(f1,font=('aria',20,'bold'),text='Samosa',width=20,fg='black',bg='Chocolate')
lbl_samosa.grid(row=3,column=0)
lbl_misal_pav=Label(f1,font=('aria',20,'bold'),text='Misal Pav',width=20,fg='black',bg='Chocolate')
lbl_misal_pav.grid(row=4,column=0)
lbl_vada_usal=Label(f1,font=('aria',20,'bold'),text='Vada Usal',width=20,fg='black',bg='Chocolate')
lbl_vada_usal.grid(row=5,column=0)
lbl_idli_sambhar=Label(f1,font=('aria',20,'bold'),text='Idli Sambhar',width=20,fg='black',bg='Chocolate')
lbl_idli_sambhar.grid(row=6,column=0)
lbl_masala_dosa=Label(f1,font=('aria',20,'bold'),text='Masala Dosa',width=20,fg='black',bg='Chocolate')
lbl_masala_dosa.grid(row=7,column=0)
lbl_bisleri=Label(f1,font=('aria',20,'bold'),text='Bisleri',width=20,fg='black',bg='Chocolate')
lbl_bisleri.grid(row=8,column=0)
lbl_pav=Label(f1,font=('aria',20,'bold'),text='Pav',width=20,fg='black',bg='Chocolate')
lbl_pav.grid(row=9,column=0)

entry_vadapav=Entry(f1,font=('aria',20,'bold'),textvariable=vadapav,bd=6,width=8,bg='coral')
entry_vadapav.grid(row=1,column=1)
entry_tea=Entry(f1,font=('aria',20,'bold'),textvariable=tea,bd=6,width=8,bg='coral')
entry_tea.grid(row=2,column=1)
entry_samosa=Entry(f1,font=('aria',20,'bold'),textvariable=samosa,bd=6,width=8,bg='coral')
entry_samosa.grid(row=3,column=1)
entry_misal_pav=Entry(f1,font=('aria',20,'bold'),textvariable=misal_pav,bd=6,width=8,bg='coral')
entry_misal_pav.grid(row=4,column=1)
entry_vada_usal=Entry(f1,font=('aria',20,'bold'),textvariable=vada_usal,bd=6,width=8,bg='coral')
entry_vada_usal.grid(row=5,column=1)
entry_idli_sambhar=Entry(f1,font=('aria',20,'bold'),textvariable=idli_sambhar,bd=6,width=8,bg='coral')
entry_idli_sambhar.grid(row=6,column=1)
entry_masala_dosa=Entry(f1,font=('aria',20,'bold'),textvariable=masala_dosa,bd=6,width=8,bg='coral')
entry_masala_dosa.grid(row=7,column=1)
entry_bisleri=Entry(f1,font=('aria',20,'bold'),textvariable=bisleri,bd=6,width=8,bg='coral')
entry_bisleri.grid(row=8,column=1)
entry_pav=Entry(f1,font=('aria',20,'bold'),textvariable=pav,bd=6,width=8,bg='coral')
entry_pav.grid(row=9,column=1)

btn_reset=Button(f1,bd=5,fg='black',bg='gold',font=('aria',15,'bold'),width=10,text='Reset',command=reset)
btn_reset.grid(row=10,column=0)

btn_total=Button(f1,bd=5,fg='black',bg='gold',font=('aria',15,'bold'),width=10,text='Total',command=total)
btn_total.grid(row=10,column=1)

billframe=Frame(root)
billframe.place(x=925,y=125)

billlabel=Label(billframe,text='Bill Area',font=('calibri',20),fg='black',bg='burlywood',width=32)
billlabel.pack()

textarea=Text(billframe,height=32,width=55,bd=5)
textarea.pack()

root.mainloop()
```
